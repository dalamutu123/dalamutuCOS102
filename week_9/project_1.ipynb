```python
import tkinter as tk
from tkinter import messagebox as msgbox

class Zenith:
    def mut_services(self):
        mutual_services = ["Lines of credit", "Investment management and accounts", "Insurance"]
        return mutual_services

    def uni_services(self):
        unique_services = ["Retirement and education accounts", "Loans and mortgages", "Checking and Saving",
                      "Multi-currency management services and products", "Foreign currency accounts", "Foreign currency credit cards",
                      "Transborder advisory services", "Liquidity Management", "Advisory services"]
        return unique_services


class glob_banking(Zenith):
    def uni_services(self):
        unique_services = ["Multi-currency management services and products", "Foreign currency accounts", "Foreign currency credit cards",
                      "Transborder advisory services"]
        return unique_services

class ret_banking(Zenith):
    def uni_services(self):
        unique_services = ["Retirement and education accounts", "Loans and mortgages", "Checking and Saving"]
        return unique_services

class com_banking(Zenith):
    def uni_services(self):
        unique_services = ["Advisory services"]
        return unique_services
    

def click():
    first_name = first_name_entry.get()
    last_name = last_name_entry.get()
    full_name = first_name + " " + last_name
    full_name = full_name.title()
    division = div_entry.get().lower()

    if division == "retail":
        r_banking = ret_banking()
        all_services = r_banking.uni_services() + r_banking.mut_services()
        msgbox.showinfo("Division Services", f"{full_name}, the services offered by your division, {division.capitalize()} Banking are: \n" + "\n".join(all_services))

    elif division == "commercial":
        c_banking = com_banking()
        all_services = c_banking.uni_services() + c_banking.mut_services()
        msgbox.showinfo("Division Services", f"{full_name}, the services offered by your division, {division.capitalize()} Banking are: \n" + "\n".join(all_services))

    elif division == "global":
        g_banking = glob_banking()
        all_services = g_banking.uni_services() + g_banking.mut_services()
        msgbox.showinfo("Division Services", f"{full_name}, the services offered by your division, {division.capitalize()} Banking are: \n" + "\n".join(all_services))

    else:
        msgbox.showerror("ERROR!", f"Division, {division} is not recognised")



# Creating the main window
root = tk.Tk()
root.title("Zenith Bank Services")
root.geometry("500x200")

# Creating label and entry for first name
first_name_label = tk.Label(root, text="First Name: ")
first_name_label.pack()
first_name_entry = tk.Entry(root)
first_name_entry.pack()

# Creating label and entry for last name
last_name_label = tk.Label(root, text="Last Name: ")
last_name_label.pack()
last_name_entry = tk.Entry(root)
last_name_entry.pack()

# Creating label and entry for division
div_label = tk.Label(root, text="Bank Division: ")
div_label.pack()
div_entry = tk.Entry(root)
div_entry.pack()

# Creating button to click for submission
submit_button = tk.Button(root, text="Check", command=click)
submit_button.pack()

# Run the main event loop
root.mainloop()

```
