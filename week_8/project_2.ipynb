```python
import tkinter as tk
from tkinter import messagebox as msgbox

class Delivery:
    prices = ["2000", "1500", "5000", "4000"]

    def __init__(self, location, weight):
        self.location = location
        self.weight = weight

    def price(self):
        if self.location == "pau" and self.weight >= 10 :
            msgbox.showinfo("Delivery Fee", f"Your delivery fee is ₦{Delivery.prices[0]}")

        elif self.location == "pau" and self.weight < 10 :
            msgbox.showinfo("Delivery Fee", f"Your delivery fee is ₦{Delivery.prices[1]}")

        elif self.location == "epe" and self.weight >= 10 :
            msgbox.showinfo("Delivery Fee", f"Your delivery fee is ₦{Delivery.prices[2]}")

        elif self.location == "epe" and self.weight < 10 :
            msgbox.showinfo("Delivery Fee", f"Your delivery fee is ₦{Delivery.prices[3]}")

        else:
            msgbox.showerror("Error", "Invalid Location")


# What happens when button is clicked
def click():
    location = location_entry.get().lower()
    weight = int(weight_entry.get())

    delivery = Delivery(location, weight)

    delivery.price()

# Creating the main window
root = tk.Tk()
root.title("Delivery Pricing App")
root.geometry("500x200")

# Creating label and entry for Location
location_label = tk.Label(root, text="Delivery Location (PAU or Epe): ")
location_label.pack()
location_entry = tk.Entry(root)
location_entry.pack()

# Creating label and entry for weight
weight_label = tk.Label(root, text="Weight of package (kg) in numbers: ")
weight_label.pack()
weight_entry = tk.Entry(root)
weight_entry.pack()

# Creating button to click for submission
submit_button = tk.Button(root, text="Check", command = click)
submit_button.pack()

# Run the main event loop
root.mainloop()
```
